```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
df2=pd.DataFrame(np.random.randint(1,100,(4,3)),index=["harsh","rolli","golu","prinshu"],columns=["DSA","Php",".net"])
df2
```

```python
x =df2.index
y =df2["Php"]
plt.subplot(1,2,1)
plt.plot(x, y,"g^-")
plt.xlabel("marks")
plt.ylabel("students")
plt.title("php marks ")
plt.grid(color='b', linestyle='--' )
plt.ylim(0,100)
plt.legend(['First line', 'Second line'])
plt.show()




x =df2.index
y =df2["DSA"]
plt.subplot(1,2,1)
plt.plot(x, y,"r", marker='*', linestyle='-',data=df2)
plt.xlabel("marks")
plt.ylabel("students")
plt.title("DSA marks ")
plt.grid(color='r', linestyle='--' )
```
